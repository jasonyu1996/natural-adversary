```python
config = []
for i in range (0, len(clean_text)):
    if(clean_text[i][0] != '-'):
        config.append(clean_text[i])
    else:
        idx = i
        break
results = clean_text[i + 1 : len(clean_text)]
info = results[-5:-1]
results = results[0:len(results) - 5]
results.append('--------------------------------')
```

```python
list_results = []
for i in range(0, len(results), 10):
    list_results.append(results[i:i+9])
```

```python
len(list_results)
```

```python
#we want only results that the model gives correct predictions
correct_pred_results = []
for i in list_results:
    if(i[0][-2] == i[6][-1]):
        correct_pred_results.append(i)
```

```python
#maybe we only want results that our adversarial examples give correct predictions from correct_pred_results
adv_correct_pred_results = []
for i in correct_pred_results:
    if(i[6][-1] == i[7][-1]):
        adv_correct_pred_results.append(i)
```
